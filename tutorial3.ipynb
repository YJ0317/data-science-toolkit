```python
import numpy as np
```

```python
array_2x2 = np.array([[1, 2], [3, 4]])
print("2x2 Array:\n", array_2x2)
```

```python
random_3x4 = np.random.rand(3, 4)
print("\n Random 3x4 Array:\n", random_3x4)
```

```python
ones_array = np.ones((2, 3))
zeros_array = np.zeros((2, 3))
print("\n Array of Ones:\n", ones_array)
print("\n Array of Zeros:\n", zeros_array)
```

```python
random_6x3 = np.random.rand(6, 3)
transposed = random_6x3.T
print("\n Random 6x3 Array:\n", random_6x3)
print("\n Transposed (3x6):\n", transposed)
```

```python
n, p, m = 3, 2, 4
A = np.random.rand(n, p)
B = np.random.rand(p, m)
dot_product = np.dot(A, B)
print("\n Matrix A (3x2):\n", A)
print("\n Matrix B (2x4):\n", B)
print("\n Dot Product (3x4):\n", dot_product)
```
